Q1. To join the 2 tables to have a common state key

```
#we need to retain GrandElectors column from Grandelectors table
#We need to retain Population column from Population table
#we should have a column 'States', common to the two tables
#Start by converting states in Electors table to UPPERCASE
#Preview the data
%%sql
UPDATE Electors
SET State = UPPER(State);
SELECT * FROM Electors LIMIT 5;
```

```
#Join the two tables to have a common column for State

%%sql
CREATE TABLE IF NOT EXISTS ElectorsPopulation AS
SELECT State,GrandElectors, Population
FROM (SELECT * FROM Electors INNER JOIN Demograph 
     ON Electors.State = Demograph.State);
SELECT * FROM ElectorsPopulation LIMIT 10
```

Q2. Change the name of the "District of Columbia" state to its short version "DC".

```
#Update the name of the record "District of Columbia" to "DC."
#Preview the table
%%sql
UPDATE ElectorsPopulation
SET State = "DC."
WHERE State = "DISTRICT OF COLUMBIA";
SELECT * FROM ElectorsPopulation LIMIT 10
```

Q4. Order the states by decreasing ratio of Grand Electors per capita. To make our priority list.

```
%%sql
UPDATE ElectorsPopulation
SET "GrandElectorsCapita" = Population/GrandElectors;
SELECT * FROM ElectorsPopulation
ORDER BY GrandElectorsCapita ASC LIMIT 10;
```
